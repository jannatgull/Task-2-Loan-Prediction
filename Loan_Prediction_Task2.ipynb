## Exploratory Data Analysis (EDA)

This section explores the distribution of important variables using
visualizations such as histograms, boxplots, and count plots to identify
patterns and trends in the data.


```python
#Visualization of data 
#Histogram is used to visualize the distribution of Annual Salary
sns.set(style= 'whitegrid')
plt.figure(figsize=(8,5))
sns.histplot(data= df , x = 'Annual Salary')
plt.title("Annual Salary Distribution")
plt.show()
```

## Model Training & Testing(Logistic Regression)

In this section, machine learning model (Logistic Regression) will be trained to predict
loan default using the selected features. The dataset will be split
into training and testing sets for evaluation.


```python
#Seperating X and Y for model training
X = df[['Index','Employed','Bank Balance','Annual Salary']]
Y = df['Defaulted?']
```

```python
#Splitting the dataset into training and testing sets
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(X , Y , train_size= 0.8 , random_state= 42)
```

```python
#Training the Logistic Regression classification model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(x_train , y_train)
#Predicting the target values using the trained Logistic Regression model
y_pred = model.predict(x_test)
```

## Evaluation Metrics (Logistic Regression)

The performance of the trained model will be evaluated using
appropriate metrics such as accuracy score and confusion matrix.


```python
from sklearn.metrics import accuracy_score , confusion_matrix
#Calculating the accuracy of the trained model
acc = accuracy_score(y_test , y_pred)
print("Accuracy score:" , round(acc , 3))
```

```python
#Creating a confusion matrix to analyze correct and incorrect predictions
cr = confusion_matrix(y_test , y_pred)
print("Classification matrix:" , cr)
```

 ## Model Training & Testing (Decision Tree)

In this section, a Decision Tree model will be trained to predict loan default using the selected features. The dataset will be split into training and testing sets to evaluate the performance of the Decision Tree.

```python
#Training the decision tree model
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train , y_train)
#Predicting the target values using the trained Decision tree model
y_pred = model.predict(x_test)
```

The performance of the trained Decision Tree model will be evaluated using appropriate metrics, including accuracy score and confusion matrix, to assess how well the model predicts loan defaults.

```python
#Calculating the accuracy of the trained model
acc = accuracy_score(y_test , y_pred)
print("Accuracy score:" , round(acc , 3))
```

```python
#Creating a confusion matrix to analyze correct and incorrect predictions
cr = confusion_matrix(y_test , y_pred)
print("Classification matrix:" , cr)
```
